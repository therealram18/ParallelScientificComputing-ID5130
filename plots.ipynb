```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = np.genfromtxt('/home/ruthwik/Downloads/data_07.csv', delimiter=',')
column1 = data[:, 0]
column2 = data[:, 1]
column3 = data[:, 2]

# Reshape each column into a 21x21 matrix
matrix1 = column1.reshape(21, 21)
matrix2 = column2.reshape(21, 21)
matrix3 = column3.reshape(21, 21)

# Plot the matrices
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
im1 = axes[0].imshow(matrix2, cmap='viridis', interpolation='nearest', extent=[0, 1, 0, 1])
axes[0].set_title('Ground Truth')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
cbar1 = plt.colorbar(im1, ax=axes[0], shrink=0.7)  # Adjust the size of the color bar
im2 = axes[1].imshow(matrix1, cmap='viridis', interpolation='nearest', extent=[0, 1, 0, 1])
axes[1].set_title('Serial Prediction')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
cbar2 = plt.colorbar(im2, ax=axes[1], shrink=0.7)  # Adjust the size of the color bar
im3 = axes[2].imshow(matrix3, cmap='viridis', interpolation='nearest', extent=[0, 1, 0, 1])
axes[2].set_title('Parallel Prediction')  # Correct the index for setting the title
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
cbar3 = plt.colorbar(im3, ax=axes[2], shrink=0.7)  # Adjust the size of the color bar
plt.tight_layout()
plt.show()
```
